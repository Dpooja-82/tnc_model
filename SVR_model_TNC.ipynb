```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr
from math import sqrt
from scipy.stats import linregress
```

```python
# Preprocess the data: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2,random_state=186)
```

```python
# Perform hyperparameter tuning using GridSearchCV
param_grid = {
    'C': [0.1, 1, 10],
    'epsilon': [0.1, 0.2, 0.5],
    'kernel': ['linear', 'rbf', 'poly']
}
grid_search = GridSearchCV(SVR(), param_grid, cv=10, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
```

```python
# Train the SVM regression model with the best hyperparameters
best_svr = SVR(**best_params)
best_svr.fit(X_train, y_train)
```

```python
# Evaluate the model's performance
y_pred = best_svr.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
pearson_corr, _ = pearsonr(y_test, y_pred)
rmse = sqrt(mean_squared_error(y_test, y_pred))


print("Evaluation Metrics:")
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
print("Pearson correlation coefficient:", pearson_corr)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
import seaborn as sns
import numpy as np
# Create a DataFrame with actual and predicted values
df_results = pd.DataFrame({'Actual_ddG_cal': y_test, 'Predicted_ddG_cal': y_pred})

# Plot actual vs. predicted values with a regression line
plt.figure(figsize=(7, 5))
# Adding jitter to the data points
jitter = 0.05
df_results['Actual_ddG_cal_jittered'] = df_results['Actual_ddG_cal'] + np.random.normal(0, jitter, size=len(df_results))
df_results['Predicted_ddG_cal_jittered'] = df_results['Predicted_ddG_cal'] + np.random.normal(0, jitter, size=len(df_results))

sns.regplot(x='Actual_ddG_cal_jittered', y='Predicted_ddG_cal_jittered', data=df_results,
            scatter_kws={'color': 'blue', 'alpha': 0.6, 's': 80}, line_kws={'color': 'gold'})

plt.text(-1.7, 1.7, r'$\mathbf{MSE}$ = %.2f'%(mse),fontsize=13, fontweight='bold')
plt.text(-1.7, 1.4, r'$\mathbf{r_{p}}$ = %.2f'%(pearson_corr), fontsize=13,fontweight='bold', )
plt.text(-1.7, 1.1, r'$\mathbf{R^{2}}$ = %.2f'%(r2),fontsize=13, fontweight='bold',)

plt.ylabel(r'$\mathbf{\Delta \Delta G_{pred}(kcal/mol})$',fontsize=16,fontweight='bold')
plt.xlabel(r'$\mathbf{\Delta \Delta G_{calc}(kcal/mol})$',fontsize=16,fontweight='bold')
plt.xticks( fontsize=11,fontweight='bold')
plt.yticks( fontsize=11,fontweight='bold')
plt.ylim(-2, 2)
plt.xlim(-2, 2)
plt.tick_params(labelsize=16)
ax = plt.subplot(111)
ax.set_facecolor('whitesmoke',)
ax.grid(True, linestyle='--', alpha=0.5)
plt.rcParams['axes.facecolor'] = 'whitesmoke'
plt.tight_layout()
plt.savefig('svm_best.png', dpi=1000)
plt.show()
```

```python
# Define SVR models with different kernels
kernels = ['linear', 'poly', 'rbf']
models = {}
predictions = {}

for kernel in kernels:
    model = SVR(kernel=kernel, C=1, epsilon=0.5)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    models[kernel] = model
    predictions[kernel] = y_pred
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr
import seaborn as sns


# Assuming kernels is a list of kernel names and predictions is a dictionary with kernel names as keys
# and their respective predictions as values

plt.figure(figsize=(15, 5))

for i, kernel in enumerate(kernels):
    plt.subplot(1, 3, i + 1)

    sns.regplot(x=y_test, y=predictions[kernel], scatter_kws={'color': 'blue', 'alpha': 0.6, 's': 80}, line_kws={'color': 'gold'})

    # Evaluate the model's performance
    y_pred = predictions[kernel]
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    pearson_corr, _ = pearsonr(y_test, y_pred)
    rmse = sqrt(mean_squared_error(y_test, y_pred))

    print(f"Evaluation Metrics for {kernel} kernel:")
    print("Mean Absolute Error:", mae)
    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
    print("Pearson correlation coefficient:", pearson_corr)
    print("Root Mean Squared Error (RMSE):", rmse)

    plt.text(-1.7, 1.7, r'$\mathbf{MSE}$ = %.2f' % (mse), fontsize=13, fontweight='bold')
    plt.text(-1.7, 1.4, r'$\mathbf{r_{p}}$ = %.2f' % (pearson_corr), fontsize=13, fontweight='bold')
    plt.text(-1.7, 1.1, r'$\mathbf{R^{2}}$ = %.2f' % (r2), fontsize=13, fontweight='bold')

    plt.xlabel(r'$\mathbf{\Delta \Delta G_{calc}(kcal/mol})$', fontsize=16, fontweight='bold')
    plt.ylabel(r'$\mathbf{\Delta \Delta G_{pred}(kcal/mol})$', fontsize=16, fontweight='bold')

    plt.xticks(fontsize=12, fontweight='bold')
    plt.yticks(fontsize=12, fontweight='bold')
    plt.ylim(-2, 2)
    plt.xlim(-2, 2)
    plt.grid(True, linestyle='--', alpha=0.5)
    ax.set_facecolor('whitesmoke',)
    plt.rcParams['axes.facecolor'] = 'whitesmoke'
plt.tight_layout()
plt.savefig('svm_kernels.png', dpi=1000)
plt.show()
```
